```python
import matplotlib.pyplot as plt

# Example data
cities = ['Paris', 'Tokyo', 'London', 'New York', 'Rome', 'Sydney']
frequency = [61, 45, 56, 78, 32,23]  
# Create bar chart
plt.bar(cities, frequency,color='purple')

# Add title and labels
plt.title('Distribution of Preferred Travel Cities')
plt.xlabel('Cities')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (age distribution)
np.random.seed(0)
ages = np.random.randint(18, 80, size=100) 

# Create histogram
plt.hist(ages, bins=10, color='pink', edgecolor='black')

# Add title and labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
